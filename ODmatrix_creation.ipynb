```python
import os
import zipfile
import pandas as pd
```

*NOTE*. I will be using OD matrix with the old data (2023). Hence, you will find more sophisticated codes under "OD matrix with the old data (2023)."

```python
def unzip_file(zip_path, extract_to):
    # Ensure the output directory exists
    os.makedirs(extract_to, exist_ok=True)

    # Open the ZIP file and extract all contents
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        # List all files in the archive
        file_list = zip_ref.namelist()
        zip_ref.extractall(extract_to)
        print(f"Extracted files to {extract_to}")
        print(f"Files extracted: {file_list}")
        return [os.path.join(extract_to, file) for file in file_list]
```

```python
# Merge ZCTA information for origins and destinations
# Rename columns
crosswalk = crosswalk.rename(columns={'tabblk2020': 'geocode', 'zcta': 'zcta'})
# Merge for origin ZCTAs: w_geocode (work)
od_data = od_data.merge(crosswalk[['geocode', 'zcta']], 
                        left_on='w_geocode', 
                        right_on='geocode', 
                        how='left').rename(columns={'zcta': 'origin_zcta'}) # It maps the w_geocode column in od_data to the corresponding zcta (ZCTA) in the crosswalk DataFrame.
# Merge for destination ZCTAs: h_geocode (home)
od_data = od_data.merge(crosswalk[['geocode', 'zcta']], 
                        left_on='h_geocode', 
                        right_on='geocode', 
                        how='left').rename(columns={'zcta': 'destination_zcta'}) # It maps the h_geocode column in od_data to the corresponding zcta (ZCTA) in the crosswalk DataFrame.
od_data
```

```python
print(od_data["origin_zcta"].isna().sum())
print(od_data["destination_zcta"].isna().sum())
```

```python
od_matrix = od_data.groupby(['origin_zcta', 'destination_zcta'])['S000'].sum().unstack(fill_value=0)
od_matrix
```

```python
# Fix above problem: origin_zcta and destination_zcta values do not match in a one-to-one relationship.
all_zcta = set(od_data['origin_zcta']).union(set(od_data['destination_zcta']))
all_zcta = sorted(all_zcta)
od_matrix = od_data.groupby(['origin_zcta', 'destination_zcta'])['S000'].sum().unstack(fill_value=0)

# Reindex rows and columns to include all ZCTAs
od_matrix = od_matrix.reindex(index=all_zcta, columns=all_zcta, fill_value=0)
print(od_matrix.shape) 
od_matrix
```
